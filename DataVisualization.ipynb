import libraries

```python
import numpy as np 
import matplotlib.pyplot as plt
import pandas as pd

data= pd.read_csv('Salary_Data.csv')
x= data.iloc[:,:-1].values
y= data.iloc[:,-1].values
```

```python

#scatterplot

plt.scatter(x,y)
plt.show()
```

```python
#lineplot

plt.plot(x,y)
plt.show()
```

```python

#histogram

plt.hist(x)
plt.show()
```

```python

#pie 
plt.pie(y)
plt.show()
```

```python

#bargraph
data.plot(kind="bar")
plt.show()
```

```python
#WhiskerBoxPlot
data.boxplot(by ='YearsExperience', column =['Salary'], grid = False)
```
